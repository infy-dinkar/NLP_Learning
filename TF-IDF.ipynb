```python
import pandas as pd
msg = pd.read_csv(r'DataSet\spam (1).csv', encoding='latin-1')
```

```python
msg.head(3)
msg=msg.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1)
```

```python
msg=msg.rename(columns={'v1':'label','v2':'message'})
```

```python
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
w_lemmatizer=WordNetLemmatizer()
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

w_lemmatizer = WordNetLemmatizer()
corpus = []

for i in range(0, len(msg)):
    # 1. Keep only letters (correct regex)
    rev = re.sub('[^a-zA-Z]', ' ', msg['message'][i])
    
    # 2. Convert to lowercase
    rev = rev.lower()
    
    # 3. Split into words
    rev = rev.split()
    
    # 4. Lemmatize + remove stopwords
    rev = [w_lemmatizer.lemmatize(word) for word in rev if word not in stopwords.words('english')]
    
    # 5. Join back into a string
    rev = ' '.join(rev)
    
    # 6. Append to corpus
    corpus.append(rev)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer(max_features=100)
X=tfidf.fit_transform(corpus).toarray()
```
